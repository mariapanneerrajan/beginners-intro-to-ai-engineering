## Topic 1: Functions and Modularity

A **function** is a reusable block of code that performs a specific task. Instead of writing the same code over and over, you write it once in a function and call it whenever you need it.

**Modularity** means breaking your code into smaller, manageable pieces. Functions are the basic building block of modularity.

### Defining and Calling Functions

Use `def` to define a function.

```python
# Simple function with no parameters
def greet():
    print("Hello, World!")

# Call the function
greet()
greet()
```

```python
# Function with parameters
def greet_person(name):
    print(f"Hello, {name}!")

greet_person("Alice")
greet_person("Bob")
greet_person("Charlie")
```

```python
# Function with multiple parameters
def add(a, b):
    result = a + b
    return result

sum1 = add(5, 3)
sum2 = add(10, 20)
print(f"5 + 3 = {sum1}")
print(f"10 + 20 = {sum2}")
```

### Return Values

Functions can return a value using `return`. This is how you get results back from a function.

```python
# Function that calculates and returns
def calculate_total(price, tax_rate):
    tax = price * tax_rate
    total = price + tax
    return total

total = calculate_total(100, 0.1)
print(f"Total: ${total}")
```

```python
# Function that returns multiple values
def get_name_parts(full_name):
    parts = full_name.split(" ")
    return parts[0], parts[1]

first, last = get_name_parts("Alice Smith")
print(f"First: {first}, Last: {last}")
```

### Default Parameters

You can provide default values for parameters.

```python
def greet_with_title(name, title="Mr."):
    print(f"Hello, {title} {name}!")

greet_with_title("Smith")  # Uses default
greet_with_title("Johnson", "Dr.")  # Overrides default
```

```python
def power(base, exponent=2):
    return base ** exponent

print(f"5 squared: {power(5)}")
print(f"5 cubed: {power(5, 3)}")
print(f"2 to the 8th: {power(2, 8)}")
```

### Scope

Variables defined inside a function are **local** to that function. They don't exist outside of it.

```python
# Variables inside functions are local
def my_function():
    local_var = "I'm inside the function"
    print(local_var)

my_function()

# This would cause an error:
try:
    print(local_var)
except NameError:
    print("Error: local_var doesn't exist outside the function")
```

```python
# Global variables (defined outside functions)
global_var = "I'm global"

def show_global():
    print(global_var)  # Can access global from inside

show_global()
print(global_var)  # Can also access from outside
```

## Topic 2: Organizing Code with Modules

A **module** is a Python file containing functions and variables. You can import modules to reuse code across files.

### Using Built-in Modules

Python comes with many useful modules. Import them with `import`.

```python
# Import the math module
import math

# Use functions from the module
print(f"Square root of 16: {math.sqrt(16)}")
print(f"Pi: {math.pi}")
print(f"Ceiling of 3.2: {math.ceil(3.2)}")
print(f"Floor of 3.8: {math.floor(3.8)}")
```

```python
# Import specific items from a module
from math import sqrt, pi

print(f"Square root of 25: {sqrt(25)}")
print(f"Pi: {pi}")
```

```python
# Import everything (use carefully)
from math import *

print(f"cos(0): {cos(0)}")
print(f"sin(pi/2): {sin(pi/2)}")
```

### Common Built-in Modules

| Module | Purpose | Example |
|--------|---------|----------|
| `math` | Mathematical functions | `math.sqrt()`, `math.pi` |
| `random` | Generate random numbers | `random.randint()`, `random.choice()` |
| `datetime` | Work with dates and times | `datetime.datetime.now()` |
| `os` | Operating system operations | `os.listdir()`, `os.path.exists()` |

```python
# The random module
import random

# Random integer between 1 and 10
number = random.randint(1, 10)
print(f"Random number: {number}")

# Random choice from a list
colors = ["red", "blue", "green", "yellow"]
chosen = random.choice(colors)
print(f"Random color: {chosen}")

# Shuffle a list
deck = [1, 2, 3, 4, 5]
random.shuffle(deck)
print(f"Shuffled: {deck}")
```

```python
# The datetime module
from datetime import datetime

now = datetime.now()
print(f"Current date and time: {now}")
print(f"Year: {now.year}")
print(f"Month: {now.month}")
print(f"Day: {now.day}")
```

### Key Takeaway

**Functions** let you:
- Write code once, use it many times
- Organize code into logical pieces
- Make code easier to test and debug
- Reuse others' code

**Modules** extend this by letting you:
- Share functions across files
- Use code written by others
- Organize large projects

Together, they're the foundation of **modularity**—breaking complex problems into smaller, manageable pieces.
