## Topic 1: The Pursuit of Higher Abstractions

Programming is about **abstracting complexity**. We write code at a high level, and the computer handles the low-level details.

Think of a car:
- **You (the driver)**: Turn the steering wheel, press pedals
- **The car's abstraction**: Handles complex mechanics, hydraulics, engines

In programming:
- **You**: Write `print("Hello")`
- **The computer**: Converts it to binary, processes it, displays it

Each programming language is an abstraction over the previous level of complexity.

## Topic 2: Binary — The Language of Computers

Everything in a computer is represented as **0s and 1s** (binary).
- `0` = OFF, no voltage
- `1` = ON, voltage

These are called **bits**. 8 bits = 1 **byte**.

### Examples:

| Decimal | Binary   | What it means |
|---------|----------|---------------|
| 0       | 00000000 | All OFF       |
| 1       | 00000001 | Last bit ON   |
| 5       | 00000101 | 4 + 1         |
| 255     | 11111111 | All ON        |

**Why learn binary?** It helps you understand that *everything* a computer does is ultimately manipulation of 0s and 1s. This is the lowest level of abstraction.

```python
# Convert decimal numbers to binary in Python
decimal_number = 5
binary = bin(decimal_number)
print(f"Decimal {decimal_number} in binary: {binary}")
print(f"Without the '0b' prefix: {binary[2:]}")
```

```python
# Convert binary back to decimal
binary_string = "101"
decimal = int(binary_string, 2)
print(f"Binary {binary_string} in decimal: {decimal}")
```

```python
# See how numbers count in binary
print("Counting 0 to 15 in binary:")
for i in range(16):
    print(f"{i:2d} → {bin(i)[2:]:>4s}")
```

## Topic 3: Intro to Python — Install and "Hello, World!"

**Python** is a high-level programming language designed to be readable and beginner-friendly.

The first program every programmer writes is `"Hello, World!"` — it's a tradition that proves the language is installed and working.

```python
# Your first Python program!
print("Hello, World!")
```

```python
# Python can do simple math
result = 2 + 3
print(f"2 + 3 = {result}")
```

```python
# And more complex operations
x = 10
y = 3
print(f"{x} + {y} = {x + y}")
print(f"{x} - {y} = {x - y}")
print(f"{x} * {y} = {x * y}")
print(f"{x} / {y} = {x / y}")
print(f"{x} // {y} = {x // y}")  # Integer division
print(f"{x} % {y} = {x % y}")    # Remainder (modulo)
```

### Key Takeaway

Binary is the foundation—everything computers do is based on 0s and 1s. But we never write in binary directly. **Python abstracts that away**, letting us write readable code that the computer translates to binary automatically. That's the beauty of higher abstractions!
