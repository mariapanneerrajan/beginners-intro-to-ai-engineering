## Topic 1: Classes and Object-Oriented Programming

**Object-Oriented Programming (OOP)** is a way of organizing code by bundling data and functions together into **objects**.

A **class** is a blueprint for creating objects. Think of it like a cookie cutter:
- **The cutter (class)**: Defines the shape
- **The cookies (objects)**: Individual instances with the same shape

```python
# Define a simple class
class Dog:
    # Constructor: runs when you create a new object
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed
    
    # Method: a function that belongs to the class
    def bark(self):
        print(f"{self.name} says: Woof!")
    
    def describe(self):
        print(f"{self.name} is a {self.breed}")

# Create objects from the class
dog1 = Dog("Buddy", "Golden Retriever")
dog2 = Dog("Max", "German Shepherd")

# Call methods on the objects
dog1.describe()
dog1.bark()

dog2.describe()
dog2.bark()
```

```python
# Classes can have multiple methods
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height
    
    def perimeter(self):
        return 2 * (self.width + self.height)
    
    def display(self):
        print(f"Rectangle: {self.width} x {self.height}")
        print(f"Area: {self.area()}")
        print(f"Perimeter: {self.perimeter()}")

rect = Rectangle(5, 10)
rect.display()
```

```python
# Classes can have attributes (data) and methods (functions)
class Bank:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.balance = balance
    
    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited ${amount}. New balance: ${self.balance}")
    
    def withdraw(self, amount):
        if amount > self.balance:
            print(f"Insufficient funds! Balance: ${self.balance}")
        else:
            self.balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.balance}")
    
    def check_balance(self):
        print(f"Account holder: {self.account_holder}")
        print(f"Balance: ${self.balance}")

# Create and use a bank account
account = Bank("Alice", 1000)
account.check_balance()
print()
account.deposit(500)
account.withdraw(200)
account.withdraw(5000)
```

## Topic 2: Working with JSON

**JSON** (JavaScript Object Notation) is a way to format data that's easy for both humans and computers to read. It's widely used for:
- Storing data in files
- Sending data over the internet
- Configuring applications

Python has a `json` module to work with JSON.

### JSON Format

JSON uses familiar structures:
- **Objects**: `{"key": "value"}` (like Python dicts)
- **Arrays**: `[1, 2, 3]` (like Python lists)
- **Values**: strings, numbers, booleans, null

```python
# JSON example
json_example = '''
{
  "name": "Alice",
  "age": 25,
  "city": "New York",
  "skills": ["Python", "JavaScript", "SQL"],
  "is_developer": true
}
'''

print("JSON text:")
print(json_example)
```

```python
# Converting JSON to Python (parsing)
import json

json_string = '{"name": "Bob", "age": 30, "city": "London"}'
python_dict = json.loads(json_string)

print(f"Type: {type(python_dict)}")
print(f"Data: {python_dict}")
print(f"Name: {python_dict['name']}")
print(f"Age: {python_dict['age']}")
```

```python
# Converting Python to JSON (serializing)
import json

person = {
    "name": "Charlie",
    "age": 28,
    "skills": ["Java", "C++"],
    "employed": True
}

json_string = json.dumps(person, indent=2)  # indent for pretty printing
print("Python object as JSON:")
print(json_string)
```

```python
# Reading JSON from a file
import json
import os

# Create a sample JSON file
sample_data = {
    "users": [
        {"id": 1, "name": "Alice"},
        {"id": 2, "name": "Bob"}
    ]
}

# Write to file
with open("sample.json", "w") as f:
    json.dump(sample_data, f, indent=2)

print("File created!")
```

```python
# Reading JSON from file
import json

with open("sample.json", "r") as f:
    data = json.load(f)

print(f"Type: {type(data)}")
print(f"Full data: {data}")
print(f"\nUsers:")
for user in data["users"]:
    print(f"  - {user['name']} (ID: {user['id']})")
```

## Topic 3: Working with APIs — First Steps

An **API** (Application Programming Interface) is a way for programs to communicate with each other over the internet.

Many APIs return data in JSON format. You can use Python's `requests` library to fetch data from APIs.

### Making HTTP Requests

The `requests` library lets you send HTTP requests to APIs.

```python
# Install requests if needed: pip install requests
# Then import it
import requests

# Example: Get a random joke from an API
url = "https://api.chucknorris.io/jokes/random"
response = requests.get(url)

print(f"Status code: {response.status_code}")  # 200 = success
print(f"Response type: {type(response.text)}")
print(f"Response:\n{response.text}")
```

```python
# Parse the JSON response
import requests
import json

url = "https://api.chucknorris.io/jokes/random"
response = requests.get(url)

# Convert JSON response to Python dict
data = response.json()

print(f"Type: {type(data)}")
print(f"Keys: {data.keys()}")
print(f"\nJoke:")
print(data["value"])
```

```python
# Example: Get weather data
import requests

# Using OpenWeatherMap API (free tier)
api_key = "demo"  # Use a real API key in practice
city = "London"
url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"

try:
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        print(f"City: {data['name']}")
        print(f"Temperature: {data['main']['temp']}°C")
        print(f"Weather: {data['weather'][0]['description']}")
    else:
        print(f"API Error: {response.status_code}")
except Exception as e:
    print(f"Error: {e}")
```

```python
# Example: A simple function to fetch and process API data
import requests

def fetch_joke():
    url = "https://api.chucknorris.io/jokes/random"
    try:
        response = requests.get(url)
        if response.status_code == 200:
            data = response.json()
            return data["value"]
        else:
            return "Failed to fetch joke"
    except Exception as e:
        return f"Error: {e}"

# Use the function
for i in range(3):
    print(f"\nJoke {i+1}:")
    print(fetch_joke())
```

### Understanding API Basics

| Term | Meaning |
|------|----------|
| **URL** | The address of the API endpoint |
| **Request** | What you ask the API to do |
| **Response** | The data the API sends back |
| **Status Code** | 200 = success, 404 = not found, 500 = server error |
| **JSON** | The format of the data returned |

### API Workflow
1. Make a request to an API URL
2. Get a response (usually JSON)
3. Parse the JSON into Python objects
4. Use the data in your program
