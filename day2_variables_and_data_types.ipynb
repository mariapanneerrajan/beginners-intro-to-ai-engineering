# Day 2: Variables and Data Types

## Topic 1: Variables and Data Types

A **variable** is a container that holds a value. Think of it like a labeled box:
- **Label**: The variable name (`x`, `name`, `age`)
- **Contents**: The value stored inside (`5`, `"Alice"`, `25`)

Every value in Python has a **data type** — it tells Python what kind of data it is and what operations you can do with it.

### The Main Data Types

| Type | Example | Description |
|------|---------|-------------|
| `int` | `5`, `42`, `-10` | Whole numbers (integers) |
| `float` | `3.14`, `2.5`, `-0.1` | Decimal numbers |
| `str` | `"hello"`, `'world'` | Text (strings) |
| `bool` | `True`, `False` | True/False values |
| `list` | `[1, 2, 3]` | Ordered collection of items |
| `dict` | `{"name": "Alice"}` | Key-value pairs |
| `None` | `None` | "No value" (represents nothing) |

```python
# Creating variables with different data types
age = 25              # int
height = 5.9          # float
name = "Alice"        # str
is_student = True     # bool

print(f"Name: {name}, Type: {type(name)}")
print(f"Age: {age}, Type: {type(age)}")
print(f"Height: {height}, Type: {type(height)}")
print(f"Is Student: {is_student}, Type: {type(is_student)}")
```

```python
# Variables can be reassigned
x = 10
print(f"x is {x}")

x = 20  # Change the value
print(f"x is now {x}")

x = "hello"  # Change the type too!
print(f"x is now {x}, type: {type(x)}")
```

## Topic 2: Working with Numbers and Strings

### Working with Numbers

Numbers support standard arithmetic operations: `+`, `-`, `*`, `/`, `//`, `%`, `**`

```python
# Arithmetic with integers
a = 10
b = 3

print(f"Addition: {a} + {b} = {a + b}")
print(f"Subtraction: {a} - {b} = {a - b}")
print(f"Multiplication: {a} * {b} = {a * b}")
print(f"Division: {a} / {b} = {a / b}")
print(f"Integer Division: {a} // {b} = {a // b}")
print(f"Remainder: {a} % {b} = {a % b}")
print(f"Exponent: {a} ** {b} = {a ** b}")
```

```python
# Working with floats (decimals)
price = 9.99
tax_rate = 0.1
tax = price * tax_rate
total = price + tax

print(f"Price: ${price}")
print(f"Tax (10%): ${tax:.2f}")     # Format to 2 decimal places
print(f"Total: ${total:.2f}")
```

### Working with Strings

Strings are sequences of characters. You can combine them, repeat them, and format them.

```python
# String concatenation (combining strings)
first_name = "Alice"
last_name = "Smith"
full_name = first_name + " " + last_name
print(f"Full name: {full_name}")
```

```python
# String repetition
laugh = "Ha"
print(laugh * 5)
```

```python
# String length
text = "Hello, World!"
print(f"Text: {text}")
print(f"Length: {len(text)} characters")
```

```python
# Accessing individual characters (indexing)
word = "Python"
print(f"First character: {word[0]}")
print(f"Third character: {word[2]}")
print(f"Last character: {word[-1]}")
```

```python
# String methods (built-in operations)
text = "hello world"
print(f"Original: {text}")
print(f"Uppercase: {text.upper()}")
print(f"Title case: {text.title()}")
print(f"Capitalize: {text.capitalize()}")
```

### Type Conversion

You can convert between types using `int()`, `str()`, `float()`, etc.

```python
# Convert string to number
age_string = "25"
age_number = int(age_string)
print(f"String: {age_string}, Type: {type(age_string)}")
print(f"Number: {age_number}, Type: {type(age_number)}")

# Now we can do math with it
print(f"Next year: {age_number + 1}")
```

```python
# Convert number to string
count = 42
count_string = str(count)
print(f"Number: {count}, Type: {type(count)}")
print(f"String: {count_string}, Type: {type(count_string)}")

# Now we can use string operations
print(f"Repeated: {count_string * 3}")
```

### Key Takeaway

Every value in Python is a specific **type**. The type determines:
- What operations you can do with it
- How it's stored in memory
- How it behaves

Variables are just labeled containers for these values. You can check a value's type with `type()` and convert between types when needed.
